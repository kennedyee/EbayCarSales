```python
#now to rename the column names
#i'll use a dictionary
new_column_dict = {'dateCrawled': 'date_crawled' , 'name' :'name' , 'seller':'seller', 'offerType':'offer_type', 'price':'price', 'abtest':'abtest',
       'vehicleType':'vehicle_type', 'yearOfRegistration':'registration_year', 'gearbox':'gearbox', 'powerPS':'power_ps', 'model':'model',
       'odometer':'odometer', 'monthOfRegistration':'registration_month', 'fuelType':'fuel_type', 'brand': 'brand',
       'notRepairedDamage':'unrepaired_damage', 'dateCreated':'date_created', 'nrOfPictures':'number_of_pictures', 'postalCode':'postal_code',
       'lastSeen':'last_seen'}
autos.rename(columns = new_column_dict, inplace= True) #columns renamed
```

```python
autos.columns #check whether that worked
autos.head()
```

```python
#now lets clean it up abit
autos.describe(include = 'all') #the include = 'all' checks for both categorical and numerical columns
```

```python
#lets change the price column to int and replace the commas and dollar $ signs
autos['price'] = (autos['price']
                        .str.replace(',','')
                        .str.replace('$','', regex=True)
                        .astype(int)
                 )
```

```python
autos['price'].dtype #check for datatype in the column
```

```python
#this removes the commas  and the km and changes the data type to int
autos['odometer'] = ( autos['odometer']
                           .str.replace('km','')
                           .str.replace(',','')
                           .astype(int)
                    )

```

```python
#lets now rename the odometer column
autos.rename({'odometer':'odometer_km'}, axis = 1, inplace = True)
```

```python
#lets see if it worked
autos[['odometer_km','price']].head()
```

```python
#lets chech for unique values
autos['price'].unique().shape
```

```python
#lets check for min,max,median,mean
autos['price'].describe()
```

```python
#now to look at the highest and lowest prices
print(autos['price'].value_counts().sort_index(ascending=True).head(30)) #gives prices in ascending order
print(autos['price'].value_counts().sort_index(ascending=False).head(10)) #gives descending order
```

```python
#to remove the incorrect data in price column and be left with range between 100 and 350,000
autos = autos[autos['price'].between(100,3500000)]
```

```python
#for the registration year column, the data type is float. 
#There are inconsistencies since we have a min year of 1000 and a max of 9999 which diesnt make sense
autos['registration_year'].describe()
```

```python
#lets solve that
#we will use the .between method
autos = autos[autos['registration_year'].between(1990,2010)]
```

```python
#scrutiny of the brand column
autos['brand'].value_counts(normalize=True)
```

```python
mean_price_by_brand ={}

for brand in autos['brand'].value_counts().head(6).index:
    
    mean_price = round(autos.loc[autos['brand'] == brand, 'price'].mean())
    
    mean_price_by_brand[brand] = mean_price
print(mean_price_by_brand)
    
```

```python
# counting mean mileage of cars by brand
mean_mileage_by_brand = {}

for brand in autos['brand'].value_counts().head(6).index:
    
    mean_mileage = round(autos.loc[autos['brand'] == brand, 'odometer_km'].mean())
    
    mean_mileage_by_brand[brand] = mean_mileage

print(mean_mileage_by_brand)
```

```python
# creating `mean_price` and `mean_mileage` dataframe to compare them
bmp_series = pd.Series(mean_price_by_brand)
bmm_series = pd.Series(mean_mileage_by_brand)

price_mileage_df = pd.DataFrame(bmp_series, columns=['mean_price'])

price_mileage_df['mean_mileage'] = bmm_series

price_mileage_df
```
